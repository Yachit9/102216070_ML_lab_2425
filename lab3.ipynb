# Binning

```python
binned_yearly_income=np.arange(min(df['YearlyIncome']),max(df['YearlyIncome']),20000)
```

```python
len(binned_yearly_income)
```

```python
YearlyIncome_label=['below low','low','Medium','above medium','High']
```

```python
df['YearlyBinned']=pd.cut(df['YearlyIncome'],bins=binned_yearly_income,labels=YearlyIncome_label)
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity


gender_m = df['Gender_M'].values.reshape(-1, 1)
marital_status_s = df['MaritalStatus_S'].values.reshape(-1, 1)


similarity = cosine_similarity(gender_m, marital_status_s)
```

```python
similarity
```

# Jaccard similarity

```python
intersection = len(set(df['Gender_M']).intersection(set(df['MaritalStatus_S'])))
union = len(set(df['Gender_M']).union(set(df['MaritalStatus_S'])))
if union == 0:
  print(0)
else:
  print( intersection / union)
```
